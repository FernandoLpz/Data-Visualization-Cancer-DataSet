# Introduction

The idea of this notebook is to show you one way to visualize a data set for a classification task. The data set is about diagnosis of cancer based on a series of features. As mentioned above the goal is to classify if the pacient has cancer or not. However in this notebook we only focus in the visualization part.

## Getting Started

```python
# Importing some libraries 
import pandas as pd # to handle the data set as frames
import seaborn as sns # to plot some visualizations
import matplotlib.pyplot as plt # to plot some visualizations
import time 

# to display inline
%matplotlib inline
```

```python
# Loading csv file
cancer_frame = pd.read_csv('./data_cancer.csv')
# Display first 5 rows
cancer_frame.head()
```

As we can see the dataset contais 33 columns or features, there are columns wich we do not need for example 'id' column, besides we need to scale the values in order to have a standardization in each column respect to each other. So we are going to drop some columns and scale the values and separate the as "target" and "corpus" wich represents the target values (malign or benign) and the rest of the corpus.

```python
# Setting 'target' as the 'y' vector and dropping columns wich are not useful
target = cancer_frame.diagnosis # as the 'Y' vector
corpus = cancer_frame.drop(['id', 'Unnamed: 32', 'diagnosis'], axis=1) # as the X matrix
```

```python
# Visualizing some features into and histogram
corpus.hist(['texture_mean'])
```

```python
# Defining and plotting a complete histogram
corpus_histogram = pd.DataFrame({
        'radius_mean':cancer_frame['radius_mean'], 
        'texture_mean':cancer_frame['texture_mean']})
corpus_histogram.plot.hist(alpha=0.5)
```

We are going to display the balance between targen benign and malign (B or M)

```python
plot_sns = sns.countplot(target, label='Count')
Benign, Malign = target.value_counts()
print('Number of Benign: ',Benign)
print('Number of Malignant : ',Malign)
```

## Scaling data

We are going to scale data using the idea of moving all values to a Gaussian Distribution with mean zero and std 1.

```python
# Scaling data, moving values to a normal distribution with mean zero and std 1
corpus_scaled = (corpus - corpus.mean()) / (corpus.std())
corpus_scaled.head()
```

## Visualizing data with different plots

Having scaled out all frame we are going to proceed to visualize the data with differents techniques and plots.

### Violin Plot

Basically the violin plot is used to visualize the distribution and the probability density of data. In this part we are goint to analize some features or colums from the frame.

```python
# Preparing data to visualize as violin plot
# Using 20 first columns

# Concatenate the target frame with just 20 columns from corpus_scaled
corpus_plot = pd.concat([target, corpus_scaled.iloc[:,0:20]], axis=1) 

# Reshaping the frame
corpus_plot = pd.melt(corpus_plot, id_vars="diagnosis", var_name="Features", value_name='Values')
corpus_plot.head()
```

```python

# Setting the plt object
plt.figure(figsize=(10,10))
# Setting the violinplot objetc with respecitve atributes
sns.violinplot(x="Features", y="Values", hue="diagnosis", data=corpus_plot, split=True, inner="quart")
# Rotation of x ticks
plt.xticks(rotation=90)
```

### Box Plot

Now lets see how data behaves using the Box Plot. 
Basically the Box Plot is used to visualize the distribution of data indeed the dispersion of them.

```python
# Visualizing first 20 elements by a box plot

# Initialize the plt objetc
plt.figure(figsize=(10,10))
# Seegin the boxplot objetc
sns.boxplot(x="Features", y="Values", hue="diagnosis", data=corpus_plot)
# Rotation of x ticks
plt.xticks(rotation=90)
```

### Marginal Plot

This kind of plot is going to show us the correlation between two variables. In this case we are going to compare 'symetry_se' and 'texture_se. Besides we are going to see the 'pearsonr' coefficient wich tell us how correlated are the variables, a coefficient more than 0.8 is said to be hard correlated.

```python
# Plotting 'joint plot' to visualize a correlation between 2 variables. 
# Correlation is taken from Pearsonr value, 1 is totally correlated.
sns.jointplot(corpus.loc[:,'symmetry_se'], 
              corpus.loc[:,'texture_se'], kind="regg", color="#ce1414")
```

### Comparing correlation between more than two variables

```python
# To compare more than two variables we are going to use different plots

# Initializing the board
sns.set(style="white")
# Setting the variables to compare
frame = corpus.loc[:,['radius_worst','perimeter_worst','area_worst']]
# Defining the object for Grid
graph = sns.PairGrid(frame, diag_sharey=False)
# Setting the kind of plot to map lower
graph.map_lower(sns.kdeplot, cmap="Reds_d")
# Setting the scatter for upper
graph.map_upper(plt.scatter)
# Setting the diagonal graph
graph.map_diag(sns.kdeplot, lw=3)
```

```python
# Initializing the board
sns.set(style="whitegrid", palette="muted")
# Initializing the plt object
plt.figure(figsize=(10,10))
# Setting atributes for plot
sns.swarmplot(x="Features", y="Values", hue="diagnosis", data=corpus_plot)
# Rotate x ticks
plt.xticks(rotation=90)
```

### Confusion Matrix

```python
# To see correlations
f, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(cancer_frame.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```
